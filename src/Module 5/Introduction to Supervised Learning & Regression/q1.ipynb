Question 1: Implementing Linear Regression <br>

Description:<br>
Implement a simple linear regression model using Python and scikit-learn to predict house prices based on one feature: the size of the house.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

house_size = np.array([600, 800, 1000, 1200, 1400]).reshape(-1, 1)  
house_price = np.array([150, 200, 250, 300, 350]) 


model = LinearRegression()
model.fit(house_size, house_price)

predicted_price = model.predict(house_size)

print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)

plt.scatter(house_size, house_price, color='blue', label='Actual Prices')
plt.plot(house_size, predicted_price, color='red', label='Regression Line')
plt.xlabel('Size of the house (sq ft)')
plt.ylabel('Price (in $1000)')
plt.title('Linear Regression: House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
