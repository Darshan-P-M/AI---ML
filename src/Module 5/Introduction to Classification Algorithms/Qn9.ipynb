Task 1: Construct and interpret a confusion matrix for tumor classification.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

# Step 1: Load dataset
data = load_breast_cancer()
X = data.data
y = data.target  # 0 = malignant, 1 = benign

# Step 2: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a classifier
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# Step 4: Make predictions
y_pred = model.predict(X_test)

# Step 5: Confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = data.target_names  # ['malignant', 'benign']

# Step 6: Plot confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Tumor Classification')
plt.show()

# Step 7: Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=labels))
```

Task 2: Create a confusion matrix for fraud detection model evaluation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Step 1: Generate synthetic fraud detection dataset (highly imbalanced)
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                           weights=[0.95, 0.05], flip_y=0, random_state=42)

# Step 2: Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 4: Predict on test set
y_pred = model.predict(X_test)

# Step 5: Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = ['Not Fraud', 'Fraud']

# Step 6: Visualize confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Fraud Detection')
plt.show()

# Step 7: Print classification report for detailed metrics
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=labels))
```


Task 3: Use confusion matrix to evaluate model predicting employee retention.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Step 1: Create synthetic dataset for employee retention (binary classification)
# 1 = Employee stayed, 0 = Employee left
X, y = make_classification(n_samples=1000, n_features=15, n_classes=2,
                           weights=[0.6, 0.4], flip_y=0.03, random_state=42)

# Step 2: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a Logistic Regression classifier
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Step 4: Make predictions on test data
y_pred = model.predict(X_test)

# Step 5: Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = ['Left', 'Stayed']

# Step 6: Plot confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Employee Retention Prediction')
plt.show()

# Step 7: Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=labels))
```
